EVEN ODD MATRIX


```python
def matrix_even_odd(matrix):
    even = []
    odd = []

    for i in range(len(matrix)):
        if i % 2 == 0:
            even.append(matrix[i])
        else:
            odd.append(matrix[i])

    even.sort()
    odd.sort()

    second_largest_even = even[-2] if len(even) > 1 else even[-1]
    second_largest_odd = odd[-2] if len(odd) > 1 else odd[-1]

    result_sum = second_largest_even + second_largest_odd

    print("Sorted even array:", " ".join(map(str, even)))
    print("Sorted odd array:", " ".join(map(str, odd)))
    print("Sum =", result_sum)

size = int(input("Enter the size of the array: "))
matrix = []

for i in range(size):
    element = int(input(f"Enter element at {i} index: "))
    matrix.append(element)

matrix_even_odd(matrix)2
4
```

Longest_palindrome_subseq


```python
def longest_palindrome_subseq(s: str) -> int:
    n = len(s)
    dp = [[0] * n for _ in range(n)]

    for i in range(n):
        dp[i][i] = 1

    for length in range(2, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            if s[i] == s[j]:
                dp[i][j] = 2 + dp[i + 1][j - 1]
            else:
                dp[i][j] = max(dp[i + 1][j], dp[i][j - 1])

    return dp[0][n - 1]

# Example usage:
s = "bbbab"
print(longest_palindrome_subseq(s))  # Output: 4
```
